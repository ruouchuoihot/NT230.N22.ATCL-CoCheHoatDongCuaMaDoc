```python
#to visualize some images from the data
def plots(ims, figsize=(20,30), rows=10, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = 10

    #to show images
    for i in range(0,50):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(list(train.class_indices.keys())[np.argmax(titles[i])], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
```

```python
#for plotting the confusin matrix
def confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names,
    )

    #to show the confusion matrix plot
    fig = plt.figure(figsize=figsize)

    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d", cmap='Blues')
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")

    #set fontsize and style of confusion matrix
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)

    #set the titles of the confusion matrix
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def Cnn():

  #define the model as sequential model
  model = Sequential()

  #layer 1  convolutional layer + relu
  model.add(Conv2D(filters = 32, kernel_size = (3,3), strides= 4, activation='relu', input_shape=(img_size,img_size,3)))

  #layer 2  convolutional layer + relu
  model.add(Conv2D(filters = 64, kernel_size = (3,3), strides= 4, activation = 'relu', input_shape=(img_size,img_size,1)))
  #layers 3 max pooling
  model.add(MaxPooling2D(pool_size = (2,2), strides = 2, padding='same'))

  #layers 4 drop out
  model.add(Dropout(0.25))

  #make a flatten layer
  model.add(Flatten())

  #layers 5  fully connected layer of 128 neurons with a ReLU activation function
  model.add(Dense(128, activation = 'relu'))

  #layer 6 set the output layer activation as a softmax
  model.add(Dense(25, activation = 'softmax'))

  #compile the model using adam optimizer
  model.compile(loss = 'categorical_crossentropy',
                optimizer = 'adam',
                metrics = ['accuracy'])

  return model
```

```python
CNN_model = Cnn()
#print the model summary
CNN_model.summary()
```

```python
history_cnn= CNN_model.fit(imgs_train, labels_train, validation_data=(imgs_val, labels_val), epochs=100)

#for plotting losses curve
loss_training = history_cnn.history['loss']
loss_test = history_cnn.history['val_loss']

#for plotting accyracies curve
accuracy_training = history_cnn.history['accuracy']
accuracy_test = history_cnn.history['val_accuracy']

loss_training = history_cnn.history['loss']
loss_test = history_cnn.history['val_loss']

accuracy_training = history_cnn.history['accuracy']
accuracy_test = history_cnn.history['val_accuracy']

#for Plotting
plt.plot(loss_test)
plt.plot(loss_training)

#set the titles of losses plot
plt.xlabel("# Of Epochs")
plt.ylabel("Loss")
plt.legend(['val Loss', 'Train Loss'])
plt.show()

#set the titles of accuracies plot
plt.plot(accuracy_test)
plt.plot(accuracy_training)
plt.xlabel("# Of Epochs")
plt.ylabel("Accuracy")
plt.legend(['val accuracy', 'Train accuracy'])
plt.show()
```

```python
#calculate and print the validation accuracy
scores = CNN_model.evaluate(imgs_val, labels_val)
print('Final validation accuraccy: ', scores[1])
```

```python
#calculate and print the testing accuracy
imgs_test, labels_test = next(test)
scores_test = CNN_model.evaluate(imgs_test, labels_test)
print('Final testing accuraccy: ', scores_test[1])
```

```python
import os
import cv2
import numpy as np

folder_a_path = "/content/drive/MyDrive/học tập/baitap/ma_doc/do_an/train"
all_descriptors = []
labels = []
# Lặp qua từng thư mục con trong folder_a
for folder_name in os.listdir(folder_a_path):
    folder_path = os.path.join(folder_a_path, folder_name)

    # Kiểm tra xem thư mục con có phải là một thư mục hay không
    if os.path.isdir(folder_path):
        # Lặp qua từng tệp tin trong thư mục con
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)

            # Kiểm tra xem tệp tin có phải là ảnh hay không
            if file_path.endswith(".jpg") or file_path.endswith(".png"):
              image = cv2.imread(file_path)
              # Khởi tạo bộ trích xuất SIFT
              sift = cv2.xfeatures2d.SIFT_create()

              step_size = 5
              kp = [cv2.KeyPoint(x, y, step_size) for y in range(0, image.shape[0], step_size)
                                              for x in range(0, image.shape[1], step_size)]

              # Trích xuất vector đặc trưng
              dense_feat = sift.compute(image, kp)
              all_descriptors.append(dense_feat)
              labels.append(file_name)
```
